# Creating a new column in a different way
Warning: Always manually check the sum of a row and compare it to the output the code calculated. There might be a miscalculation.


---


sum(axis=1) means add the values in row (horizontal addition).
sum(axis=0) means add the values in column (vertical addition).


---


```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1) # First : means "Do this calculation to every row". Second : specifies which columns we want to add together.
# Warning: 4:10 means include column index 4 up to column index 9 (BUT NOT includind column index 10)
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]] # rearranging columns
```

```python
df.to_csv('modified.csv', index=False) # modifying to a new csv file and not including the index column
df.to_excel('modified.xlsx', index=False) # modifying to a new xlsx file and not including the index column
df.to_csv('modified.txt', index=False, sep="\t") # modifying to a new \t seperated text file and not including the index column
```

# **Filtering Data**

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # finding rows with 'Type 1' is 'Grass' and 'Type 2'is 'Poison.
# Note: Generally in python, we use (and), (or) but in the pandas library we use the &, |, etc. symbols like in above line.
new_df # Notice that the indices stayed the same and didn't reset

new_df.reset_index(drop=True, inplace=True) # resets index then gets rid of the 'Index' column. If we don't want to assign it again to the same var, we can do it inplace

df.loc[df['Name'].str.contains('Mega')] # filters rows that contain the word Mega in the Name column

df.loc[~df['Name'].str.contains('Mega')] # filters rows that DO NOT contain the word Mega in the Name column

```

```python
import re # import regular expressions

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] # filter rows that have either fire or grass in the Type 1 column while ignoring caps bec we put flags=re.I.
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)] # filter rows with word that have pi + any letter after it that are zero or more (bec we have *)
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)] # ^ means it must start with what comes after
```

# **Conditional Changes**

# Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')

df.groupby(['Type 1']).mean() # gets stats of all pokemons by Type 1 and gives you the mean of the other columns acc to each Type 1

df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)

# To make things easier and see the count of every pokemon, create a new column called 'count'
df = pd.read_csv('modified.csv')

df['count'] = 1
df.groupby(['Type 1']).count()['count'] # shows count column only. Counts how many Pokemons of each Type 1

df.groupby(['Type 1','Type 2']).count()['count'] # squeezing the data frame and making it smaller but counting also acc to Type 1 and Type 2 at the same time
```
